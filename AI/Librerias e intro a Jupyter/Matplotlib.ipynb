## Representación gráfica de datos
Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el indice de cada punto de datos en el array como la coordenada horizontal.

```python
# Representación de gráfico lineal
lista = [1, 2, 5, 4, 7, 8, 3, 1]

# A la función "plot" se le pasa la lista. La lista se mostrará en el eje Vertical
plt.plot(lista)
# "show" muestra el grafico
plt.show()
# En el eje horizontal no se le ingresa nada porque lo saca automáticamente. Nada mas
# representa de manera secuencial lo que le ingresamos como lista
```

```python
# También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.
# Esto se hace proporcionando dos listas. Si ya tengo la "lista" definida en el paso anterior, defino
# una segunda lista. Hay que tener en cuenta que TANTO "X", COMO "Y" TIENEN QUE TENER LA MISMA CANTIDAD
# DE ELEMENTOS. Deben ser del mismo tamañp
lista2 = [1, 4, 6, 7, 8, 9, 10, 15]
plt.plot(lista2, lista)
plt.show()
```

```python
# Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
# La tabla se ve muy ajustada a los bordes, entonces para darle mas márgenes se le agrega el
# método "axis" como se muestra en la línea siguiente
```

```python
# Para definir más márgenes en el gráfico
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
# Le decimos que el gráfico vaya entre las coordenadas que le indicamos
# Desde "-4" en el eje Horizontal hasta "8" y en el eje Vertical va desde "0" hasta "10"
plt.axis([-4, 8, 0, 10])
plt.show()
```

```python
# Se sigue el mismo procedimiento para pintar una función matemática
import numpy as np

# Se obtiene un array con "linspace" que vaya desde "-2", hasta "2" y que vayan "500" valores
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
# Tambien se puede modificar el estilo de la gráfica para que contenga más información

plt.plot(x, y)
# Para agregarle titulo
plt.title("Square function")
plt.xlabel("X")
plt.ylabel("y 0 x**")
plt.grid(True) # Agrega cuadrícula
plt.show()
```

```python
# Pueden superponerse gráficas y cambiar el estilo de las líneas
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

# Para reresentar una grafica, encima de la otra
#plt.plot(x, y, x, y2) # Esta es la forma básica de representarlas

# Pero si se le quiere dar estilos, se puede agregar.
# De los primeros argumentos "x, y, 'b--'," indica que la primera función va a ser Azul y con
# líneas discontínuas.
# La segunda función va a ser Verde y con línea contínua, los argumentos que lo definen son "x, y2, 'g'"
plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
# Otras formas de hacer lo mismo con otras sentencias
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

# Se acumulan dentro de la misma gráfica
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt. show()
```

```python
# Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

# label="x**2"": Agrega una leyenda
plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")

# plt.legend(loc="best"): Es para que ubique la leyenda en el mejor lugar que considere y donde no tape tanto la gráfica
plt.legend(loc="best")
plt. show()
```

```python
# También se puede crear dos gráficas que no se superpongan. Estas gráficas se organizan en un grid y se denominan subplots.
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

# Con subplot se indica: el número de filas que va tener este subplot que es "1", la cantidad de columnas que es "2"
# y la posición que va a ocupar en este grid que es "1".
# Esto lo que hace es crear un grid con 1 fila, 2 columnas y lo siguiente que me saque por pantalla (que me haga "plot")
# lo voy a localizar en la primera posición, en la primera casilla de éste grid
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')

# Despues vuelvo a invocar el subplot y le digo que en éste mismo grid (que tiene 1 fila y 2 columnas) tiene que localizar
# éste nuevo "plot" en la segunda casilla
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g')

plt. show()
```

```python
# Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande. Para hacerlos, se usa "figure"
# Esto es para que cuando inicialice la figura, inicie con la siguiente configuración
# "figsize()"
plt.figure(figsize=(14, 6))

x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize=14) # Se agregan los labels pero con fuente tambien modificada
plt.ylabel("y1", fontsize=14)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g')

plt. show()
```

## Scatter plots

```python
# Vamos a pintar gráficas que tengan puntos, lo que se llama "Scatter plots"
# Primero se van a definir puntos aleatorios
from numpy.random import rand

# Se obtienen valores aleatorios entre 2 y 100
x, y = rand(2, 100)
```

```python
# Representación gráfica
plt.scatter(x, y)
plt.show()
```

```python
# Tambien se pueden superponer
x, y = rand(2, 100)
x2, y2 = rand(2, 100)

plt.scatter(x, y, c='red') # Colorea con rojo
plt.scatter(x2, y2, c='blue')
plt.show()
```

## Histograma

```python
# Para el histograma se define una lista
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
# Un histograma tiene barras que suben hasta un valor determinado
# Subplot tiene un sintaxis que permite poner los parametros todos juntos como si fuesen un número entero (SIN ESTAR SEPRADOS POR COMA)
# Este ejemplo tiene 2 filas, 1 columna y está en la primera grilla
plt.subplot(211)

# Para hacer el histograma se le pasan los datos, "bins" que son el numero de barras que quiero que haya. También se le puede
# pasar el ancho de las barras con "rwidth"
plt.hist(data, bins=15, rwidth=0.8)
plt.xlabel("value")
plt.ylabel("Frequency")

plt.subplot(212)
plt.hist(data, bins=5, rwidth=0.8)
plt.xlabel("value")
plt.ylabel("Frequency")

plt.show()
```
