# Sobre el entorno

Se pueden incrustar celdas de texto en el código. En local con "Anaconda" se hace seleccionando la opción **Markdown** y en el entorno "online" se hace clickeando la opción **"+ Texto"** y para el código **"+ Código"**.
Luego se ejecuta lacelda con **Ctrl + F9**.

## Atajos útiles de Teclado
- "Esc" : Salir de una celda
- "J" y "K": Son para moverse arriba y abajo
- "Enter": Para entrar a una celda
- "A": Si se está por fuera de la celda, crea una nueva celda
- "Ctrl + M + D": Borra la celda selleccionada
- "Shift + Enter": Ejecuta la celda y abre una nueva abajo
